```python
import pandas as pd

s= pd.read_csv("Student_Marks.csv")
data=pd.DataFrame(s)
data
```

```python
import pyttsx3
from sklearn.linear_model import LinearRegression
X=data[["time_study"]]
y=data[["Marks"]]
model = LinearRegression()
model.fit(X,y)


predicted_score = model.predict(X)
print("predicted_score is",predicted_score)
```

```python
from sklearn.metrics import mean_absolute_error 

X=data[["time_study"]]
MAE = mean_absolute_error(X,predicted_score)
print("MAE = ",MAE)
```

```python
from sklearn.metrics import mean_squared_error
X=data[["time_study"]]
MSE = mean_squared_error(X,predicted_score)
print("MSE = ",MSE)
```

```python
from sklearn.metrics import mean_squared_error
import numpy as np
s=np.sqrt(mean_squared_error(X,predicted_score))
print("MSE:",s)
```
